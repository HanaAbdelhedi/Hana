#Explanation of the Efficinet Frontier above: 

That was the efficient frontier. We want to be in the green above dot. But it's very hard to be there 
as long as we are estimating the expected returns which can be missleading. 
So instead we try to hit and reach the yellow or the  blue point. The yellow is the equally weighted portfolio point
and the blue is the global minimim variance portfolio. 
Advantages: No expected return required  for either  the blue or the yellow dot. That is called the naive portfolio. 
The Yellow dot is quite simple to reach but the blue one is a lot harder to reach as we will need an estimate for the covariance matrix(  but still is it a lot easier to estimate than the expected return) which makes it  a more reasonable target than the one base on estimation of the expected return( green dot).
